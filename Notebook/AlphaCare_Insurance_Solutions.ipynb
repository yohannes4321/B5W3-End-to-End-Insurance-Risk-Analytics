```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('..\data\MachineLearningRating_v3.txt',sep='|')
df.head()
```

```python
df.info()
```

Cheack null value and fillna and dropna values based on wheater they are important or not 

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df['Gender'].fillna(df['Gender'].mode())
df['VehicleType'].fillna(df['VehicleType'].mean())
```

```python
df.dropna()
```

```python
df.nunique()
```

```python
df_clean = df[(df['TotalPremium'] > 0) & (df['TotalPremium'].notna()) & (df['TotalClaims'].notna())]
df_clean['LossRatio'] = df_clean['TotalClaims'] / df_clean['TotalPremium']
data = ['Province', 'VehicleType', 'Gender']
plt.figure(figsize=(12, 10))

for idx, d in enumerate(data, 1):
    plt.subplot(len(data), 1, idx)
    grouped = df_clean.groupby(d)['LossRatio'].mean().sort_values(ascending=False)
    
    sns.barplot(x=grouped.index, y=grouped.values, palette='viridis')
    plt.title(f'Average Loss Ratio by {d}')
    plt.xlabel(d)
    plt.ylabel('Average Loss Ratio')
    plt.xticks(rotation=45)
    plt.tight_layout()

plt.show()
```

```python
df_data=df[(df['TotalPremium'] >0) & (df['TotalPremium'].notna()) & (df['TotalClaims']).notna()]
df['loss_data']=df_data['TotalClaims']/df_data['TotalPremium']
plt.figure(figsize=(13,8))
data = ['Province', 'VehicleType', 'Gender']
for idx,va in enumerate(data,1):
    
    plt.subplot(len(data),1,idx)
    g=df_data.groupby(va)['loss_data'].mean().sort_values(ascending=False)
    sns.barplot(x=g.index,y=g.values)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.ylabel('loss_data')
    plt.xlabel(va)
plt.show()

```
