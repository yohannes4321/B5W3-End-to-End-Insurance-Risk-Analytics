```python
x=df.iloc[:,3:13]
y=df.iloc[:,13].values

```

```python
x['Geography']=x['Geography'].astype('category')
x['Gender']=x['Gender'].astype('category')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

Converting Dataset into Dmatrix 

```python
import xgboost as xgb
```

```python
xgb_train=xgb.DMatrix(x_train,y_train,enable_categorical=True) 
xgb_test=xgb.DMatrix(x_test,y_test,enable_categorical=True)
```

```python
parameter={
    'objective':'binary:logistic',#binary for logistic 
    'max_depth':3,
    'learning_rate':0.1,

}
n=50
model=xgb.train(params=parameter,dtrain=xgb_train,num_boost_round=n)
```

```python
preds=model.predict(xgb_test)
from sklearn.metrics import accuracy_score,classification_report
preds = np.round(preds)

accuracy=accuracy_score(y_test,preds)
con_mat=classification_report(y_test,preds)
print(f' Accuracy {accuracy}')
print(f' confustion matrix {con_mat}')
```

```python
x.shape
y.shape
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=1)
```

```python
import xgboost as xgb

# Create regression matrices
dtrain_reg = xgb.DMatrix(X_train, y_train, enable_categorical=True)
dtest_reg = xgb.DMatrix(X_test, y_test, enable_categorical=True)
```

```python
params={'objective':'reg:squarederror',
        'max_depth':3}
model=xgb.train(params=params,dtrain=dtrain_reg,num_boost_round=100)
pred=model.predict(dtest_reg)
pred=np.round(pred)
print(accuracy_score(y_test,pred))
```
